#### Using Matplotlib to make pretty graphs

```python
#Simple line plot
t = arange(0.0,1.01,0.01)
c = cos(4*pi*t)
#Note: pi comes loaded, so it does not have to be definted in the python notebok
#Note: pi can be redefined, so be careful
#Can be fixed by using pi=np.pi
s = sin(4*pi*t)
#Printing s outputs the array
```

```python
plot(t,s)
#typing in show() in a new line will get rid of the text above the plot
```

```python
#How to add axis labels
plot(t,s)
xlabel("Time (s)",fontsize=14)
ylabel("Voltage (mV)",fontsize=14)
title("My Cool Plot",fontsize=20)
#Chaning the font/whatnot can be done in the line
show()
```

```python
#To show both graphs on same plot
plot(t,s,color="blue", linewidth=3)#How to change the color and linewidth
plot(t,c,color="red", linewidth=3) 
xlabel("Time (s)",fontsize=14)
ylabel("Voltage (mV)",fontsize=14)
title("My Cool Plot",fontsize=20)
show()
```

```python
plot(t,s,color="blue", linewidth=3, label="$sin(4{\pi}t)$") #How to add in a legend to the graphs
plot(t,c,color="red", linewidth=3, label="$cos(4{\pi}t)$") #How to get the mathy/Greek symbols
xlabel("Time (s)",fontsize=14)
ylabel("Voltage (mV)",fontsize=14)
title("My Cool Plot",fontsize=20)
legend()
show()
#ex: if we want uppercase sigma, we'd type \Sigma.
#similarly, lowercase sigma would be \igma
```

```python
plot(t,s,color="blue", linewidth=3, label="$sin(4{\pi}t)$") 
plot(t,c,color="red", linewidth=3, label="$cos(4{\pi}t)$") 
xlabel("Time (s)",fontsize=14)
ylabel("Voltage (mV)",fontsize=14)
title("My Cool Plot",fontsize=20)
legend()
grid(True) #How to add gridlines
savefig("simple_plot.pdf") #Saves the file to the working directory you're in, in the format of a supported filetype
show()
```
