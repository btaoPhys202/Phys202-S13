Electric Dipole Potential
-

$V(x,y)=\frac{kq}{\sqrt{x^2+(y-d)^2}}-\frac{kq}{\sqrt{x^2+(y+d)^2}}$

```python
dx = 0.06 #Define step sizes as to avoid infinities. Using a step size of 0.05 will skew the graph
dy = 0.06
x = arange(-2.,2.,dx)
y = arange(-2.,2.,dy)
X,Y = meshgrid(x,y)
print X.ndim
print Y.shape
print X
print Y
```

```python
#Make the function, V(x,y)
k = 1; q = 1; d = 1
V=(k*q/(X**2+(Y-d)**2)**(1/2.))-(k*q/(X**2+(Y+d)**2)**(1/2.))
print V
```

```python
#additional pieces for 3D plotting
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.spectral)
#help(Axes3D.plot_surface)
cb = fig.colorbar(p, shrink=0.5) #Creating a color bar neeeds to be assigned to the plot
```

Using Contours to Represent 3D Data
-

```python
#Need to define a set of array values
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])

contour(X,Y,V,levels,linewidths=4,cmap=cm.prism)
show()
```

Determine the electric potential for a quadrupole consisting of $a$ = $-1.0$ nC at both ($x$ = $\pm2$, $y$ = $0$), and $ q$ = $+1.0$ nC at both ( $ x $ = $0$, $y$ = $\pm2 $ ). Make a surface plot and a contour plot on the range -4< $ x,y $ <4 using 101 bins in $ x $ and 101 bins in $ y $. What is the value of the function at $V(20,30)$    ?

```python
dx = 4/(101.)
dy = 4/(101.)
x = arange(-4.,4.,dx)
y = arange(-4.,4.,dy)
X,Y = meshgrid(x,y)

k = (8.987551787)*(10**9) #(N*m**2)/(C**2)
q = e**(-9) #Coulombs
d = 2 #meters
V = -(k*q/(X**2+(Y+d)**2)**(1/2.))-(k*q/(X**2+(Y-d)**2)**(1/2.))+(k*q/(Y**2+(X-d)**2)**(1/2.))+(k*q/(Y**2+(X+d)**2)**(1/2.))

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.spectral)
```

```python
def pointPotential(x,y,q,posx,posy):
    """Takes point(s), the size and direction of charge, and the position of that charge to return the electric potential due to that charge"""
    k = (8.987551787)*(10**9)
    Vxy = (k*q)/(((x-posx)**2)+((y-posy)**2))**(1/2.)
    return Vxy
```

```python
x = 1.; y = 1.
V11 = pointPotential(x,y,1e-9,0.,0.)
print V11
```

```python
x,y = meshgrid(arange(-2.,2.,0.06),arange(-2.,2.,0.06)) 
Vxy = pointPotential(x,y,1e-9,0.,0.) 
print Vxy
```

```python
def dipolePotential(x,y,q,d):
    """Takes point(s), the size and direction of charge, and the position of that charge to return the dipole potential due to that charge"""
    Vxy = pointPotential(x,y,q,-d,0.)-pointPotential(x,y,q,d,0.)
    return Vxy
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x, y, dipolePotential(x,y,1,1), rstride=1, cstride=1, linewidth=0, cmap=cm.spectral)
```
