Electric Dipole Potential

$V(x,y)=\frac{kq}{\sqrt{x^2+(y-d)^2}}-\frac{kq}{\sqrt{x^2+(y+d)^2}}$

```python
dx = 0.06 #Define step sizes as to avoid infinities. Using a step size of 0.05 will skew the graph
dy = 0.06
x = arange(-2.,2.,dx)
y = arange(-2.,2.,dy)
X,Y = meshgrid(x,y)
print X.ndim
print Y.shape
print X
print Y
```

```python
#Make the function, V(x,y)
k = 1; q = 1; d = 1
V=(k*q/(X**2+(Y-d)**2)**(1/2.))-(k*q/(X**2+(Y+d)**2)**(1/2.))
print V
```

```python
#additional pieces for 3D plotting
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.spectral)
#help(Axes3D.plot_surface)
cb = fig.colorbar(p, shrink=0.5) #Creating a color bar neeeds to be assigned to the plot
```

Using Contours to Represent 3D Data

```python
#Need to define a set of array values
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])

contour(X,Y,V,levels,linewidths=4,cmap=cm.prism)
show()
```

Exercise 1: Make a 3D plot of the function $G(x,y)=e^{(-x^2-y^2)}$ over the range -6<$ (x,y) $<6 using 51 bins in $x$ and 51 bins in $y$.

```python
dx = 12/(51.)
dy = 12/(51.)
x = arange(-6.,6.,dx)
y = arange(-6.,6.,dy)
X,Y = meshgrid(x,y)

print Y.shape
```
