```python
x,y = meshgrid(arange(-0.05,0.0501,0.0011),arange(-0.05,0.0501,0.0011))
q1 = -2.e-9 #Coulombs
xp = 0.; yp = -0.005 #position of q1 in meters
qd = 3.e-9 #Coulombs
d = 0.01 #dipole separation in meters
Vtotal = pointPotential(x,y,q1,xp,yp) + dipolePotential(x,y,qd,d)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
p = ax.plot_surface(x,y,Vtotal,rstride = 1, cstride=1,linewidth=0,cmap=cm.GnBu)
ylabel('y(m)')
xlabel('x(m)')
show()
```

```python
levels = array([-20000.,-10000.,-5000.,-2000.,-1000.,-500.,-250.,0.,250.,500.,1000.,2000.,5000.,10000.,20000.])
contour(x,y,Vtotal,levels,cmap=cm.hot)
ylabel('y(m)')
xlabel('x(m)')
show()
```
