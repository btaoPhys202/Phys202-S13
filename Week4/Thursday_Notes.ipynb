Using Boolean Masks for Finding Data
-

```python
theta = arange(0.,2*pi,pi/100) #Create an array
plot(theta,sin(theta))
show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)
print mask #array of original array, but in boolean terms
```

```python
plot(theta,sin(theta),theta[mask],sin(theta[mask])) #Can plot multiple things on the same graph.
show()
```

```python
y = sin(theta[mask])
jj = (y == max(y))
print jj #size is smaller than original array, but still a boolean mask
```

```python
#Want to set values of where y is true to NaN (not a number [computer's way of handling infinity])
y[jj] = NaN
plot(theta, sin(theta))
plot(theta[mask],y,linewidth=4)
show()
```


Interpolation
-

```python
theta = arange(0.,2*pi+0.01,pi/10)
mask = (theta == 0.5)
#We are only definining the function at certain points. Interpolation lets us find values between those poitns.
print mask
```

```python
newTheta = 0.5; newSin = 0.
from scipy.interpolate import interp1d #file on the filesystem somewhere that interpolates.
#help(interp1d)
sinInterp = interp1d(theta,sin(theta))
newSin = sinInterp(newTheta) #Use function to get values where we want them
print newSin
plot(theta,sin(theta),newTheta,newSin,'rx',markersize=20,markeredgewidth=3)
show()
```

```python
#To get more values
thetaNew = arange(0,2*pi+0.1,pi/30)
sinNew = interp1d(theta, sin(theta))(thetaNew)
plot(theta, sin(theta), thetaNew, sinNew, 'r.')
show()
```

```python
plot(thetaNew, sinNew-sin(thetaNew)) #Plot the interpolated values vs. actual sine values
ylabel("Difference between interp and exact values")
show()
```

```python
#Cubic spline interpolation, using a polynomial
sinNewC = interp1d(theta, sin(theta), kind='cubic')(thetaNew)
subplot(2,1,1)
plot(thetaNew,sinNew-sin(thetaNew),thetaNew,sinNewC-sin(thetaNew),'r') 
subplot(2,1,2)
plot(thetaNew,sinNewC-sin(thetaNew),'r')
show()
```

```python
x = arange(0.,10.01,0.1)
y = x**5
```

```python
dy = diff(y) #y(n+1)-y(n)
#help(diff)
print len(y), len(dy) #Notice that dy is shorter.
dx = diff(x)
sydy = dy/dx
print len(dy/dx)
```
