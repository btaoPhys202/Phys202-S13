```python
t = arange(-4, 4+0.1, 0.03)
f = (t+1)/(t-1)
g = exp(-t)

plot(t,f, color="blue", label="(1+t)/(t-1)")
plot(t,g, color="red", linestyle="--", label="exp(-t)")

xlabel("time (seconds)", fontsize=18)
ylabel("f(t),g(t)", fontsize=18)
xticks(arange(-4,4+0.1,2), ("-4","-2","0","2","4"))
yticks(arange(-10,10+0.1,5), ("-10","-5","0","5","10"))

xmin, xmax = xlim(-4,4)
ymin, ymax = ylim(-10,10)

legend()
show()
```
