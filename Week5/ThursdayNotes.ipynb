Example 1

Let's plot with the function $f(x)=x^{1/2} + \sin(x) $ 
to see what we are working with.

```python
f = lambda x: x**(0.5) + sin(x)
```

```python
x = arange(0.,15.,0.001)
```

```python
myplot = plot(x,f(x),lw=3)
ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
xlabel(r'$x$',fontsize=20)
xlim(0.,15.)
show()
```

So we want to integrate between the limits or (1.7,12.4). We will create a bounding box for our random sampling from these limits and the maximum value of the function within that range.

```python
xmin = 1.7
xmax = 12.4

#find the maximum value of f within the boundary to set ymax.
subx = x[logical_and((x>xmin),(x<xmax))]
ymax=max(f(subx))*1.05
ymin=0.

#Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

#plot the box
plot([xmin,xmin],[ymin,ymax],color='k',linestyle='--') #vertical left
plot([xmax,xmax],[ymin,ymax],color='k',linestyle='--') #vertical right
plot([xmin,xmax],[ymax,ymax],color='k',linestyle='--') #horizontal top

show()
```

Now, we'll create random sample points within the boundary. We'll create a set of sample points along the x and y axes. The random_sample() function returns random numbers within the bounds [0.,1.]. We need to scale and translate the values into our range.

```python
#random_sample?

N = 10000
samples_x = xmin + (xmax-xmin)*random_sample(N);
samples_y = ymin + (ymax-ymin)*random_sample(N);

#Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:200],samples_y[1:200],'.')

#plot the box
plot([xmin,xmin],[ymin,ymax],color='k',linestyle='--') #vertical left
plot([xmax,xmax],[ymin,ymax],color='k',linestyle='--') #vertical right
plot([xmin,xmax],[ymax,ymax],color='k',linestyle='--') #horizontal top

show()
```

The fraction of points $(x_i,y_i)$ that satisy the confition $y_i<=f(x_i)$ is an estimate of the ratio of the integral of f(x) to the area of the rectangle.

```python
newmask = (samples_y < f(samples_x))
```

```python
sum(newmask)
```

Now we can replot the points that lie below the function.

```python
#Replot the function
myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

plot(samples_x[1:200],samples_y[1:200],'.')

#Show the ones below the function in red
plot(samples_x[newmask[:200]],samples_y[newmask[:200]],'r.')

#plot the box
plot([xmin,xmin],[ymin,ymax],color='k',linestyle='--') #vertical left
plot([xmax,xmax],[ymin,ymax],color='k',linestyle='--') #vertical right
plot([xmin,xmax],[ymax,ymax],color='k',linestyle='--') #horizontal top

show()
```

Now we can compute our integral from step V: the area of the rectangle times the ratio of the number of red and green dots.

```python
I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size
print I
```

$\int f(x)dx = \int x^{1/2}dx + \int sin(x)dx = \frac{2}{3} x^{3/2} - cos(x)$

```python
Iana = lambda x: (2./3.)*x**(3./2.)-cos(x)
Iexact = Iana(xmax)-Iana(xmin)
Ierr = 100*(Iexact-I)/Iexact
print "I_exact =", Iexact
print "I_hitormiss =", I
print "Error on I =", Ierr, "%"
```
